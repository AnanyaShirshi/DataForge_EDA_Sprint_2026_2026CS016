##Problem Statement:
Raw data loaded into Power BI is not sufficient for meaningful analysis without defined calculations and relationships. The task was to create a structured data model and write DAX (Data Analysis Expressions) measures and calculated columns to enable dynamic, aggregated analysis across the dataset.


---

---





##Procedure:
. IDE Setup: Power BI.

. Workflow:

  1. Opened Power BI Desktop → clicked the Model View icon on the left sidebar to inspect the table structure and relationships.
  2. Navigated to Report View → right-clicked the dataset table in the Fields pane → selected New Measure → wrote the Total Sales DAX measure.
  3. Created a second measure using New Measure → wrote the Profit Margin % formula using the DIVIDE function to safely handle divide-by-zero cases.
  4. Right-clicked the dataset table → selected New Column → wrote the Shipping Delay calculated column using DATEDIFF between Order Date and Ship Date.
  5. Dragged the Total Sales measure onto the report canvas as a Card visual to verify it returned a correct aggregated value.
  6. Took a screenshot of the Model View showing the table schema → saved as model_schema.png.
  7. Copied all three DAX formulas into a text file → saved as dax_formulas.txt.

##Observations:

Q1. What is the difference between a column and a measure?

Ans:

1. Column: Stores a value for every row and is a part of the dataset
2. Measure: Performs dynamic calculations based on the current filters and is mainly used to summarize data in reports.

##Conclusion:


DAX measures and calculated columns transform a flat data table into a dynamic analytical model. By separating row-level calculations (calculated columns) from aggregation logic (measures), Power BI can respond to any filter combination instantly — making DAX the core engine that powers interactive dashboards
