##Problem Statement:
Visual tools like Excel and Power BI have limitations when handling large datasets programmatically. The task was to perform a comprehensive Exploratory Data Analysis (EDA) using Python in Google Colab — examining dataset structure, computing descriptive statistics, detecting outliers, and generating visualizations entirely through code to enable reproducible, scalable analysis.


---

---





##Procedure:
. IDE Setup: Google Colab.

. Workflow:

  1. Opened Google Colab → created a new notebook → renamed it 07_inspection.ipynb
  2. Uploaded dataset_cleaned.xlsx using files.upload() → loaded it into a Pandas DataFrame using pd.read_excel() with the openpyxl engine.
  3. Inspected dataset structure using df.shape, df.columns.tolist(), df.dtypes, and df.isnull().sum() to understand dimensions, column names, data types, and missing values.
  4.Generated descriptive statistics using df.describe() to obtain count, mean, standard deviation, min, quartiles, and max for all numeric columns.
  5. Plotted a Sales Distribution histogram using sns.histplot() to examine the frequency distribution and skewness of the Sales column.
  6. Plotted a Profit by Category boxplot using sns.boxplot() to compare profit spread and outliers across the three product categories.
  7. Generated a Correlation Heatmap using sns.heatmap() on numeric columns to identify relationships between Sales, Profit, Quantity, Discount, and Shipping Cost.
  8.Applied IQR-based outlier detection on the Profit column — calculated Q1, Q3, IQR, lower and upper bounds — and printed the count and bounds of detected outliers.
