##Problem Statement:
Raw data and DAX measures alone do not communicate insights to stakeholders. The task was to design and build an interactive Power BI dashboard that visually presents key business metrics from the Global Superstore dataset, enabling non-technical users to explore sales, profit, and operational performance through dynamic charts and slicers.


---

---





##Procedure:
. IDE Setup: Power BI.

. Workflow:

  1. Opened Power BI Desktop → Report View → enabled Gridlines and Snap to Grid under the View tab for neat visual alignment.
  2. Added a Clustered Bar Chart → assigned Region to the Y-axis and Total Sales (DAX measure) to the X-axis to show sales performance by region.
  3. Added a Line Chart → assigned Order Date (grouped by Month) to the X-axis and Profit to the Y-axis to show profit trend over time.
  4. Added a Donut Chart → assigned Category to Legend and Total Sales to Values to show category-wise sales breakdown.
  5. Added two Card visuals at the top of the canvas — one for Total Sales and one for Profit Margin % — to display headline KPI numbers.
  6. Added a Slicer visual → assigned Order Date grouped by Year → set style to Tile from the Format pane so users can click a year to filter all visuals simultaneously.
  7. Applied a consistent Theme from the View tab, added titles to each visual, and arranged all visuals into a structured layout — KPI cards on top, charts below, slicer in the top right.
  8. Saved as superstore_dashboard.pbix → exported to PDF via File → Export → Export to PDF.
