##Problem Statement:
Descriptive EDA tells us what happened in the data, but not why or what will happen next. The task was to build a machine learning classification model that predicts whether a given order will be profitable or result in a loss, based on features such as Sales, Quantity, Discount, Shipping Cost, Segment, Category, Region, and Shipping Delay — enabling proactive business decision-making.


---

---





##Procedure:
. IDE Setup: Google Colab.

. Workflow:

  1. Engineered two new features from existing columns — Profit Margin (Profit divided by Sales) and Shipping Delay (difference in days between Ship Date and Order Date using .dt.days), after converting both date columns to datetime using pd.to_datetime().
  2. Encoded categorical columns (Segment, Category, Region) into numeric values using LabelEncoder from scikit-learn, creating three new encoded columns.
  3. Created the binary target variable Target — assigned 1 if Profit > 0 (Profitable) and 0 otherwise (Loss) using .astype(int).
  4. Selected 8 features (Sales, Quantity, Discount, Shipping Cost, Segment_encoded, Category_encoded, Region_encoded, Shipping Delay) as the feature matrix X and Target as y.
  5. Split the data into training and test sets using train_test_split() with 80% train and 20% test, random_state=42.
  6. Trained a RandomForestClassifier with 100 estimators on the training set using model.fit(X_train, y_train).
  7. Evaluated the model using classification_report() to obtain precision, recall, and F1-score, and plotted a Confusion Matrix using sns.heatmap().
  8. Plotted Feature Importance as a horizontal bar chart and saved the trained model as superstore_model.pkl using joblib.dump().
