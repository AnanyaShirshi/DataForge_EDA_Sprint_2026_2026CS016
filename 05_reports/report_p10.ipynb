##Problem Statement:
Analytical insights locked inside Colab notebooks are not accessible to business stakeholders. The task was to deploy a fully interactive web dashboard built with Streamlit that allows any user to explore the Global Superstore dataset through dynamic filters and visualizations — bridging the gap between data analysis and real-world decision-making by making the EDA publicly accessible via a live URL.


---

---





##Procedure:
. IDE Setup: Google Colab.

. Workflow:

  1. Installed required libraries in Colab using !pip install streamlit pyngrok -q.
  2. Created app.py using %%writefile app.py — built a multi-section Streamlit dashboard with sidebar filters (Region, Category), four KPI metrics (Total Sales, Total Profit, Average Discount, Average Shipping Delay), and four charts (Sales by Region bar chart, Profit by Category boxplot, Sales Trend line chart, Correlation Heatmap).
  3. Loaded the dataset inside the app using @st.cache_data decorator on the load_data() function to cache the data and prevent reloading on every user interaction.
  4. Implemented sidebar multiselect filters for Region and Category that dynamically filter the entire dashboard on selection.
  6. Authenticated ngrok by running ngrok.set_auth_token() with a free account token from the ngrok dashboard.
  7. Launched the Streamlit server in the background using subprocess.Popen() on port 8501 with --server.headless=true flag.
  8. Created a public tunnel using ngrok.connect(8501) — obtained a live public URL accessible from any browser.
  9. Took a screenshot of the running app → saved as app_screenshot.png → downloaded app.py using files.download() for submission.
