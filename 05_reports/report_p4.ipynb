##Problem Statement:
Raw datasets often contain inconsistency such as mixed data types, errors and composite fields that are not ready for analysis. Build a reproducible ETL (Extract Transform Load) pipeline in Power BI's Power Query Editor that cleans, restructures, and enriches the Global Superstore dataset before it reaches the visualization or modeling layer.


---

---





##Procedure:
. IDE Setup: Power BI.

. Workflow:

  1. Opened Power BI Desktop → Home → Get Data → Text/CSV → selected dataset_raw.xlsx -> clicked Transform Data to open the Power Query Editor.
  2. Applied Use First Row as Headers to promote the first row as column names.
  3. Identified columns with errors -> right clicked -> Replace Errors -> replaced with null to prevent downstream failures.
  4. Selected the Order ID column -> Home -> Split Column -> By Delimiter -> used as delimiter -> split at each occurrence, producing three sub-columns (Region Code, Year, Numeric ID).
  5. Navigated to Add Column → Conditional Column → created a new column named Profitability: If Profit > 0 -> "Profitable", Else -> "Loss".
  6. Opened View -> Advanced Editor -> copied the full auto-generated M code -> saved as power_query_M_code.txt.
  7. Clicked Close & Apply to load the transformed dataset into the Power BI data model.

##Observations:

Q1. How is Power Query different from manual Excel cleaning?

Ans: Power Query is faster, more reliable, and reusable, making it the preferred choice for cleaning large or recurring datasets. Manual Excel cleaning is suitable for small, one-time tasks, but it is time-consuming and more prone to errors when data needs frequent updates.

---



Q2. Which transformation would save time when new CSV data arrives?

Ans:Power Query transformations save time when new CSV data arrives because the cleaning steps are recorded and can be automatically reapplied by simply refreshing the query. This ensures consistency, reduces manual effort, and minimizes errors.
