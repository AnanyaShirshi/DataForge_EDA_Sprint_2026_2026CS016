##Problem Statement:
Not all customers behave the same way — some are high-value repeat buyers while others are discount-driven occasional shoppers. The task was to apply unsupervised machine learning (K-Means clustering) to group customers into distinct segments based on their purchasing behavior, enabling targeted marketing, personalized pricing, and strategic customer relationship management.


---

---





##Procedure:
. IDE Setup: Google Colab.

. Workflow:

  1. Aggregated the dataset at the customer level using df.groupby('Customer ID') — computed total Sales, total Profit, total Quantity, mean Discount, total Shipping Cost, and order count (Count of Order ID) per customer, resulting in one row per unique customer.
  2.Renamed the aggregated Order ID column to Order Count for clarity using .rename().
  3. Selected 6 clustering features (Sales, Profit, Quantity, Discount, Shipping Cost, Order Count) and applied StandardScaler to normalize all features to the same scale, preventing high-value columns like Sales from dominating the distance calculations.
  4. Applied the Elbow Method — ran K-Means for K values from 2 to 10, recorded inertia (within-cluster sum of squares) for each, and plotted an Elbow Curve to identify the optimal number of clusters.
  5.Identified K=4 as the optimal cluster count from the elbow plot where inertia reduction begins to flatten significantly.
  6. Fitted the final KMeans model with n_clusters=4, random_state=42, n_init=10 → assigned cluster labels to each customer using fit_predict().
  7. Visualized clusters on a Scatter Plot of Total Sales vs Total Profit, colored by cluster label using plt.scatter() with cmap='Set1'.
  8. Profiled each cluster by computing mean feature values per cluster using groupby('Cluster').mean() → interpreted each cluster's business meaning → saved the clustered data as customer_segments.csv.
